```python

from sklearn.tree import DecisionTreeClassifier

# Load the Titanic dataset (you need to have a dataset or download it)
# For this example, let's assume you have a CSV file named 'train.csv'
# The dataset should contain columns like 'Pclass', 'Age', 'Sex', 'Fare', 'Survived', etc.

# Load the dataset
data = pd.read_csv('/content/Titanic_train.csv')

# Preprocessing: Handling missing values and converting categorical variables to numerical
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

# Select features and target
features = ['Pclass', 'Age', 'Sex', 'Fare']
target = 'Survived'

X = data[features]
y = data[target]

# Create a Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(X, y)

# Take input from the user
print("Enter passenger details:")
pclass = int(input("Pclass (1/2/3): "))
age = float(input("Age: "))
sex = int(input("Sex (0 for male, 1 for female): "))
fare = float(input("Fare: "))

# Make a prediction
example_passenger = [[pclass, age, sex, fare]]
prediction_result = clf.predict(example_passenger)

if prediction_result[0] == 1:
    print("Prediction: Survived")
else:
    print("Prediction: Did not survive")
```
